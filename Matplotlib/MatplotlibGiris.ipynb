```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
yasListesi = [10,20,30,30,30,40,50,60,70,75]
yasNumpyArray = np.array(yasListesi)
```

```python
kiloListesi = [20,30,40,40,50,60,65,70,80,90]
kiloNumpyArray = np.array(kiloListesi)
```

```python
plt.plot(yasNumpyArray,kiloNumpyArray,"r") #ilk girdi x ekseni 2. girdi y ekseni olur. "r" ise rengi belirtir "b" -> mavi "g" -> yeşil  
plt.xlabel("Yaş") # x eksenin altında isim gösterir
plt.ylabel("Kilo") # y eksenin altında isim gösterir
plt.title("Kilonun Yaşa Göre Değişim Grafiği") # grafiğin üstünde başlık gösterir
plt.show() #grafiği gösterir
```

```python
numpyDizisi1 = np.linspace(0,20,10)
numpyDizisi2 = numpyDizisi1 ** 3
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"g+-") # -- -> kesik çizgi *- -> verilerin bulunduğu yerlere * koyar  
plt.show()
```

```python
plt.subplot(1,2,1) # subplot -> yan yana veya altta grafikler çizerken kullanılır. 1 sıra, 2 kolon olacağını ve ilk grafiği çiziyoruz
plt.plot(numpyDizisi1,numpyDizisi2,"r--")
plt.subplot(1,2,2) # 1 sıra ve 2 kolon olacağını belirttik ve 2. kolondaki veriyi çiziyoruz
plt.plot(numpyDizisi2,numpyDizisi1,"b*-")
plt.show()
```

## Figure

```python
myFigure = plt.figure() # figure -> kendi grafiğimizi oluşturmamızı sağlar
figureAxis = myFigure.add_axes([0.2,0.2,0.9,0.9]) #x ve y ekseni ekler.(ilk 2 sayı altta açıklanıyor) width ve hight ayarı yapılır(son 2)
figureAxis.plot(numpyDizisi1,numpyDizisi2,"r") # grafik oluştu
figureAxis.set_xlabel("X verisi") #x eksenine isim verdik
figureAxis.set_ylabel("Y verisi") #y eksenine isim verdik
figureAxis.set_title("Grafik Başlığı") #başlığımıza isim verdik

```

```python
figure2 = plt.figure()

eksen1 = figure2.add_axes([0.1,0.1,0.9,0.9]) #ilk sayı sol tarafta nerede durduğunu, 2. ise üst tarafta nerede durduğunu ayarlıyor
eksen2 = figure2.add_axes([0.2,0.6,0.3,0.3])

eksen1.plot(numpyDizisi1,numpyDizisi2,"r")
eksen1.set_xlabel("X ekseni")
eksen1.set_ylabel("Y ekseni")
eksen1.set_title("Grafik Başlığı")

eksen2.plot(numpyDizisi2,numpyDizisi1,"b--")
eksen2.set_xlabel("Alt X Ekseni")
eksen2.set_ylabel("Alt Y ekseni")
eksen2.set_title("Alt Grafik Başlığı")

#aynı figure'den farklı grafikler oluşturabilir bu da iç içe grafik olarak gözükür
```

## Subplots

```python
(benimFigure, benimEksen) = plt.subplots() #figure ve ekseni oto oluşturur.
benimEksen.plot(numpyDizisi1,numpyDizisi2,"r")
benimEksen.set_xlabel("X ekseni")
#benimFigure -> grafiği gösterir
```

```python
(benimFigure1, benimEksen1) = plt.subplots(nrows=1,ncols=2) #1 satır ve 2 kolon oalcağını belirttik (oto 2 figure ve eksen oluştu)
#benimEksen1.plot(numpyDizisi1,numpyDizisi2,"r") -> bunu yapamayız çünkü artık eksenimiz numpy dizisi içerisinde for kullanlacak

for eksen in benimEksen1:
    eksen.plot(numpyDizisi1,numpyDizisi2,"r")
    eksen.set_xlabel("X ekseni")
#for döngüsü ile dizimizdeki eksenleri grafik yaptık

plt.tight_layout() #grafiklerin iç içe girmesini engeller
#benimFigure1
```

```python
yeniFigure = plt.figure()
#figsize(3,3) -> tablo boyutunu ayarlar
#dpi=150 -> çözünürlülüğü ayarlar

yeniEksen = yeniFigure.add_axes([0.1,0.1,0.7,0.7])
yeniEksen.plot(numpyDizisi1,numpyDizisi2,label="İlk veri") #label -> veri çizgisi ile ilgili kıas bilgi verir
yeniEksen.plot(numpyDizisi2,numpyDizisi1,label="Son veri")
yeniEksen.legend() # label'leri tablo içinde gösterir
```
