```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
numpyDizisi1 = np.linspace(0,10,20)
numpyDizisi2 = numpyDizisi1 ** 2
```

```python
(benimFigure, benimEksen) = plt.subplots()
benimEksen.plot(numpyDizisi1,numpyDizisi2,"r")
benimEksen.plot(numpyDizisi2,numpyDizisi1, color="#28B3D8", alpha=0.3) #renk kodları ile çizgiler renlendirilebilir alpha -> saydamlaştırır
```

```python
(yeniFigure, yeniEksen) = plt.subplots(dpi=150)
yeniEksen.plot(numpyDizisi1, numpyDizisi1+2,"r",linewidth=5.0) #çizgiyi kalınlaştırır
yeniEksen.plot(numpyDizisi1,numpyDizisi1+3,"b", linestyle="--") #Çizgiye şekil verir
yeniEksen.plot(numpyDizisi1,numpyDizisi1+4,"g", linestyle="-.")
yeniEksen.plot(numpyDizisi1,numpyDizisi1+5,"y", linestyle=":")

yeniEksen.plot(numpyDizisi1,numpyDizisi1+6,color="#000000",marker="o", markersize=10.0,markerfacecolor="red")
#marker -> verilerin olduğu yerlere sembl koyar, markersize -> boyutunu ayarlar, markerfacecolor -> rengini ayarlar

```

## scatter

```python
plt.scatter(numpyDizisi1,numpyDizisi2) #verilerin bulunduğu yerleri gösterir
```
