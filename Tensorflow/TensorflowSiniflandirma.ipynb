## modeli oluşturma

```python
y = dataFrame["Type"].values
x = dataFrame.drop("Type", axis=1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=10)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train = scaler.transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
model = Sequential()

model.add(Dense(30, activation="relu")) # modellerde kaç kolon varsa o kadar nöron eklenmesi tavsiye edilir
model.add(Dense(15, activation="relu")) # sonraki katmanlar yarısı olabilir
model.add(Dense(15, activation="relu"))


model.add(Dense(1, activation="sigmoid")) #0 ve 1 arasında değer almak istiyoruz bu yüzden çıkış katmanına fonkisyon verdik
model.compile(optimizer="adam", loss="binary_crossentropy")
```

```python
model.fit(x=x_train, y=y_train, epochs=700, validation_data=(x_test,y_test), verbose=1)
```

```python
kayipVerisi = pd.DataFrame(model.history.history)
```

```python
kayipVerisi.head(3)
```

```python
kayipVerisi.plot()
```

## Modeli erken durdurma

```python
model = Sequential()

model.add(Dense(30, activation="relu")) 
model.add(Dense(15, activation="relu")) 
model.add(Dense(15, activation="relu"))


model.add(Dense(1, activation="sigmoid"))
model.compile(optimizer="adam", loss="binary_crossentropy")
```

```python
earlyStopping = EarlyStopping(monitor="val_loss", mode="min", verbose=1, patience=25)
#monitor -> kontrol edilecek veriyi seçeriz
#mode -> minimum veya maksimum değerin mi bulunacağını ayarlar
#patience -> eğer model öğrenimi kötü gidiyorsa yazılan epochs'ta durdurur
```

```python
model.fit(x=x_train, y=y_train, epochs=700, validation_data=(x_test,y_test), verbose=1, callbacks=[earlyStopping])
```

```python
modelKaybi = pd.DataFrame(model.history.history)
```

```python
modelKaybi.plot() # earlystopping, val_loss artmaya başladığında model eğitimini durdurdu
```

```python
model = Sequential()

model.add(Dense(30, activation="relu")) 
model.add(Dropout(0.6)) # verileri rastgele alıp silmeyi dener ve doğruluğu arttırır

model.add(Dense(15, activation="relu")) 
model.add(Dropout(0.6))

model.add(Dense(15, activation="relu"))
model.add(Dropout(0.6))

model.add(Dense(1, activation="sigmoid")) 
model.compile(optimizer="adam", loss="binary_crossentropy")
```

```python
model.fit(x=x_train, y=y_train, epochs=700, validation_data=(x_test,y_test), verbose=1, callbacks=[earlyStopping])
```

```python
modelKayibi = pd.DataFrame(model.history.history)
```

```python
modelKayibi.plot()
```
