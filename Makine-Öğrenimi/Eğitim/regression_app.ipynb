```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
dataFrame = pd.read_csv("biyolojiDataset.csv")
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
linear_reg = LinearRegression()
```

```python
x = dataFrame.pelvic_incidence.values.reshape(-1,1)
y = dataFrame.sacral_slope.values.reshape(-1,1)
```

```python
plt.scatter(x,y)
plt.xlabel("Pelvic")
plt.ylabel("Sacral")
```

```python
linear_reg.fit(x,y)
```

```python
y_head = linear_reg.predict(x)
```

```python
plt.scatter(x,y)
plt.plot(x, y_head, "r")
plt.xlabel("Pelvic")
plt.ylabel("Sacral")
```

```python
linear_reg.intercept_ #bias, sabit
```

```python
linear_reg.coef_ #ağırlık
```

```python
# formül ile yapım
a = linear_reg.coef_
b = linear_reg.intercept_

print("Tahmin: ", a * 50 + b) # 50 -> pelvic(feature)
```

```python
# tahmin yolu ile yapım
linear_reg.predict([[50]]) 
```

```python
from sklearn.metrics import r2_score
```

```python
print("Performans: ", r2_score(y, y_head))
```
